Discrete Uniform PMF

```python
import numpy as np
import matplotlib.pyplot as plt

def discrete(n):
    x = np.arange(1, n + 1)
    value = 1/n
    y = np.full_like(x,value,dtype=np.double)
    plt.bar(x, y, label=f'n={n}')
    plt.title('Discrete Uniform PMF')
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.xticks(np.arange(1, n + 1, 1))
    plt.legend()
    plt.show()
x=int(input())
discrete(x)
```

Bernoulli PMF

```python
import numpy as np
import matplotlib.pyplot as plt

def bernoulli(p):
    x = np.arange(2)
    y=(p**x)*((1-p)**(1-x))
    plt.bar(x, y,width=0.2)
    plt.title('Bernoulli PMF')
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.show()
x=float(input())
bernoulli(x)
```

Binomial PMF

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def binom(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))

def binomial(n, p):
    k = np.arange(0, n + 1)
    q = [(p ** i) * ((1 - p) ** (n - i)) * binom(n, i) for i in k]
    plt.bar(k, q)
    plt.title('Binomial PMF')
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.show()

x = int(input())
y = float(input())

binomial(x, y)
```

Poisson PMF

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def poisson(lam,n):
    x = np.arange(1, n + 1)
    val = np.exp(-lam) * (lam**x) / [math.factorial(k) for k in x]
    plt.bar(x, val)
    plt.title('Poisson PMF')
    plt.xlabel('Number of Events')
    plt.ylabel('Probability')
    plt.show()
lam=int(input())
n=int(input())
poisson(lam,n)
```

Geometric PMF

```python
def geometric(p, k):
    x = np.arange(1, k + 1)
    val = [(1-p)**(k-1) * p for k in x]
    plt.bar(x, val, width=0.5)
    plt.title('Geometric PMF')
    plt.xlabel('Number of Trials')
    plt.ylabel('Probability')
    plt.show()
p=float(input())
k=int(input())
geometric(p,k)
```

Equivalence of Binomial and Poisson for large values of M and small values of p

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def binom(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n - k))

def binomial(n, p):
    k = np.arange(0, n + 1)
    q = [(p ** i) * ((1 - p) ** (n - i)) * binom(n, i) for i in k]
    plt.bar(k, q)
    plt.title('Binomial PMF')
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.show()

def poisson(lam,n):
    x = np.arange(1, n + 1)
    val = np.exp(-lam) * (lam**x) / [math.factorial(k) for k in x]
    plt.bar(x, val)
    plt.title('Poisson PMF')
    plt.xlabel('Number of Events')
    plt.ylabel('Probability')
    plt.show()
n=int(input())
x=int(input())
y=float(input())
lam=n*y
binomial(x,y)
poisson(lam,n)
```
