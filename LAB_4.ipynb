```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

PDF

Uniform

```python
a = int(input())
b = int(input())
x =np.arange(a-1, b+1, 0.001)
def uniform(x, a, b):
    val = np.zeros(len(x))
    for i in range(len(x)):
        if a <= x[i] <= b:
            val[i] = 1 / (b - a)
        else:
            val[i] = 0
    return val
y = uniform(x, a, b)
plt.plot(x, y)
plt.fill_between(x, y, alpha=0.5)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('PDF of Continuous Uniform Distribution')
plt.grid(True)
plt.show()
```

Gaussian

```python
def gaussian(x, mu, sigma):
    return (1 / (sigma * np.sqrt(2 * math.pi))) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))
u1 = float(input())
s1 = float(input())
u2 = float(input())
s2 = float(input())
x = np.arange(u1 - 4*s1, u1 + 4*s1, 0.001)
y1 = gaussian(x, u1, s1)
y2 = gaussian(x, u2, s2)
plt.plot(x, y1, label=f'Gaussian PDF (μ1={u1}, σ1={s1})')
plt.plot(x, y2, label=f'Gaussian PDF (μ2={u2}, σ2={s2})')
plt.fill_between(x, y1, alpha=0.2)
plt.fill_between(x, y2, alpha=0.2)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title('PDF of Gaussian Distributions')
plt.legend()
plt.grid(True)
plt.show()
```

Cauchy

```python
def cauchy(val, x, gamma):
    val = (1 / (math.pi * gamma)) * (gamma**2 / ((val - x)**2 + gamma**2))
    return val

x1 = float(input())
g1 = float(input())
x2 = float(input())
g2 = float(input())
x3 = float(input())
g3 = float(input())
x4 = float(input())
g4 = float(input())

val1 = np.arange(x1 - 10*g1, x1 + 10*g1, 0.001)
y1 = cauchy(val1, x1, g1)
val2 = np.arange(x2 - 10*g2, x2 + 10*g2, 0.001)
y2 = cauchy(val2, x2, g2)
val3 = np.arange(x3 - 10*g3, x3 + 10*g3, 0.001)
y3 = cauchy(val3, x3, g3)
val4 = np.arange(x4 - 10*g4, x4 + 10*g4, 0.001)
y4 = cauchy(val4, x4, g4)

plt.plot(val1, y1, label=f'Cauchy PDF (x1={x1}, γ1={g1})')
plt.fill_between(val1, y1, alpha=0.2)
plt.plot(val2, y2, label=f'Cauchy PDF (x2={x2}, γ2={g2})')
plt.fill_between(val2, y2, alpha=0.2)
plt.plot(val3, y3, label=f'Cauchy PDF (x3={x3}, γ3={g3})')
plt.fill_between(val3, y3, alpha=0.2)
plt.plot(val4, y4, label=f'Cauchy PDF (x4={x4}, γ4={g4})')
plt.fill_between(val4, y4, alpha=0.2)

plt.xlabel('x')
plt.ylabel('PDF')
plt.title('PDF of Cauchy Distributions')
plt.legend()
plt.grid(True)
plt.show()
```

CDF

Uniform

```python
def uniform(x, a, b):
    if x < a:
        return 0
    else:
        return (x - a) / (b - a)

a=int(input())
b=int(input())
x = np.arange(a, b, 0.001)
cdf = [uniform(i, a, b) for i in x]

plt.plot(x, cdf, label='CDF')
plt.fill_between(x, cdf, alpha=0.2)
plt.title('CDF of Uniform Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def uniform_pdf(x, a, b):
    return 1 / (b - a) if a <= x <= b else 0

def uniform_cdf(x, a, b):
    if x < a:
        return 0
    else:
        return (x - a) / (b - a)

a = int(input())
b = int(input())
x = np.arange(a, b, 0.001)
y = [uniform_pdf(xi, a, b) for xi in x]
analytical = [uniform_cdf(xi, a, b) for xi in x]
cumsum = np.cumsum(y) * 0.001

plt.plot(x, analytical, label='Analytical CDF')
plt.plot(x, cumsum, label='Cumulative Sum CDF', linestyle='--')


plt.title('CDF of Uniform Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

Gaussian

```python
def gaussian_pdf(x, mean, deviation):
    factor = 1 / (deviation * np.sqrt(2 * np.pi))
    return factor * np.exp(-((x - mean) ** 2) / (2 * deviation ** 2))

m = float(input())
d = float(input())
m2 = float(input())
d2 = float(input())

x = np.arange(min(m, m2) - 4*max(d, d2), max(m, m2) + 4*max(d, d2), 0.001)
pdf = gaussian_pdf(x, m, d)
pdf2 = gaussian_pdf(x, m2, d2)

dx = x[1] - x[0]
cdf = np.cumsum(pdf) * dx
cdf2 = np.cumsum(pdf2) * dx

plt.plot(x, cdf, label=f'CDF μ={m}, σ={d}')
plt.plot(x, cdf2, label=f'CDF μ={m2}, σ={d2}')
plt.fill_between(x, cdf, alpha=0.2)
plt.fill_between(x, cdf2, alpha=0.2)
plt.title('CDF of Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def cauchy_pdf(x, x0, gamma):
    return 1 / (np.pi * gamma * (1 + ((x - x0) / gamma)**2))

def cauchy_cdf(x, x0, gamma):
    return 0.5 + (1 / np.pi) * np.arctan((x - x0) / gamma)

x0 = float(input())
gamma = float(input())

min = x0 - 10 * gamma
max = x0 + 10 * gamma
x = np.arange(min, max, 0.001)

pdf = cauchy_pdf(x, x0, gamma)
analytical = cauchy_cdf(x, x0, gamma)

delta_x = x[1] - x[0]
cumsum = np.cumsum(pdf) * delta_x

plt.plot(x, analytical, label='Analytical CDF', color='blue')
plt.plot(x, cumsum, label='Cumulative Sum CDF', linestyle='--')

plt.title('Validation of CDF of Cauchy Distribution')
plt.xlabel('x')
plt.ylabel('CDF')
plt.legend()
plt.grid(True)
plt.show()
```

Bernoulli

```python
p=float(input())

x = np.array([0, 1])
y = np.cumsum(np.array([1-p, p]))

y=y/np.sum(y)
cdf = np.cumsum(y)
plt.step(x,cdf)
plt.xlabel("X")
plt.ylabel("CDF")
plt.title("CDF of Bernoulli Variable")
plt.grid(True)
plt.show()
```

Binomial

```python
n = int(input())
p = float(input())
x = np.arange(0, n+1)
pmf = np.array([math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in x])
cdf = np.cumsum(pmf)
plt.step(x, cdf, where='post')
plt.xlabel('X')
plt.ylabel('CDF')
plt.title('CDF of Binomial Distribution')
plt.grid(True)
plt.show()
```

Poisson

```python
lambd = float(input())
n=int(input())
x = np.arange(0,n+1 )
pmf = np.array([np.exp(-lambd) * (lambd**k) / math.factorial(k) for k in x])
cdf = np.cumsum(pmf)
plt.step(x, cdf, where='post')
plt.xlabel('X')
plt.ylabel('CDF')
plt.title('CDF of Poisson Distribution')
plt.grid(True)
plt.show()
```
